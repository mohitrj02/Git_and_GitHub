#### Add & Commit
- add : adds new or changed files in your working directory to the Git staging area.

    git add<-file name->

- commit : it is the record of changes

    git commit -m"some message"

**Add :**
1. what ever changes maid or new file added, it is transfered to the "Staging Area", means ready to be commited.
2. "git add file_name" this will add the file to staging area
3. if we have multiple files to add, use "git add ." this will add all the files to the staging area.
5. Now files are ready to commit.

**Note : Only the "commits" are recorded in the Git not the "Add", "clone", etc.**

**Commit :**
1. "git commit -m 'some msg'", this will commit the changes added
2. Let's check the "git status"

    **Your branch is ahead of 'origin/main' by 1 commit.**
3. Now, we have commited the changes in our local system but these are not commited on the Git yet. For this we need "Push Command"

### Init Command
- init : used to create a new git repository


```python
# git init
# git remote add origin<-link->
# git remote -v  (to verify remote)
# git branch (to check branch)
# git branch -M main (to rename branch)
# git push origin main
```

**Note : we calls the GitHub repository as gits remote repository.**

1. First, we creates a new directory / repository using the "mkdir".
2. The new created repo is a local repo, so we transform it as a Git repo using the "git init" command.
3. then, we works and creates new file and add to staging area using "git add ." and then, "git commit -m 'msg'" to commit the adds.

    But, these are not done on the Git yet. To do so, we do few steps:
    1. create a new repository in the GitHub manually without a Readme file. Bcz, if we creates it, then we need to take that to the local system also. Copy the "https: link"
    2. Add new remote repo using "git remote add origin<-link->". Here, origin is basically the reference for the "link". Here, we meant that we wanted to add a new git remote repo and wanted to call it as "Origin".
    3. Now, "git remote -v" it basically, checks that The git remote repo which we are talking about is exactly what!.
    4. "git branch", refers to the different branches of the repo. Names the branch in which we are working.
    5. "git branch -M main" : to rename the branch bcz earlier the "master" was the default branch for GitHub but, now the "main" is the default branch. So, either create a new "main" branch & work on that or, rename the "master" as "main".

4. Now, we push all the changes using "git push origin main".

**We can also use the command "git push -u origin main" where, the "-u" is basically means to **set upstream**. Means, if we are working in a single repo for long time and don't wanted to write origin main everytime so, we use "-u flag" to say that all the push are for that single origin main only and we can only write as "git push" afterward for next pushes.**

### Undoing Changes
**When commited something mistakenly, then to reverse that and do 'undo' changes**

Case 1 : staged changes (which are added but not commited)

    git reset <-file_name-> : for a single file
    git reset               : for multiple files

Case 2 : commited changes (for one commit)

    git reset HEAD~1

**The latest commit it named as "HEAD" by default. HEAD~1 means to reset he HEAD to the last previous commit.**

Case 3 : commited changes (for many commits)

    git reset <-commit_hash->
    git reset --hard <-commit_hash->

**Hash is basically the 'code' for a perticular 'commit'. Which we can easily get using the 'git log' and we gets log of commit with their details**
